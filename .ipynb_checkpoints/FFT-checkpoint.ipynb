```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
fp = 'Liu_processed/1.png'
img_gray = cv2.imread(fp, cv2.IMREAD_GRAYSCALE)
ret, img_binary = cv2.threshold(img_gray, 127, 1, cv2.THRESH_BINARY_INV)
```

```python
f = np.fft.fft2(img_binary)
fshift = np.fft.fftshift(f)
s1 = np.log(np.abs(fshift))
n = s1.shape[0]
s2 = s1[int(n/2):int(n/2)+50,int(n/2):int(n/2)+50]
plt.figure()
plt.imshow(s2)
plt.show()
plt.imshow(s1)
plt.show()
```
