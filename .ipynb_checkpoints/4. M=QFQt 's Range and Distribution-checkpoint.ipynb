# The Range of Matrix M = QFQt for All Characters

```python
dir_frames = 'Liu_processed'
filenames = []
filesinfo = os.scandir(dir_frames)
filenames = [f.path for f in filesinfo if f.name.endswith(".png")]
filenames.sort(key=lambda f: int(''.join(filter(str.isdigit, f))))
fileCount = len(filenames)
print(fileCount)
```

```python
fp = './Liu_processed/1.png'
img_gray = cv2.imread(fp, cv2.IMREAD_GRAYSCALE)
ret, img_binary = cv2.threshold(img_gray, 127, 1, cv2.THRESH_BINARY_INV)
k =img_gray.shape[1]
k
```

```python
def i_to_x(i):
    global k
    return (2*i-k+1)/(k-1)
def j_to_y(j):
    global k
    return (2*j-k+1)/(k-1)

# the function for calculating H_p_tilde(i,k;sigma)
def discrete_hermite(p,i,sigma):
    x = i_to_x(i)
    return ((2**p)*factorial(p)*math.sqrt(math.pi)*sigma)**(-1/2)*math.exp(-x**2 / (2*sigma**2))*sp.special.eval_hermite(p, x/sigma)
    
    
def discrete_hermite_Q(n,k,sigma):
    Q = np.zeros((n+1,k), dtype="double")
    for i in range(n+1):
        for j in range(k):
            Q[i,j] = discrete_hermite(i,j,sigma)
    return Q
```

```python
def M_range(k, n, filenames, fileCount):
    M_max = np.zeros((n+1,n+1), dtype="double")
    M_mix = np.zeros((n+1,n+1), dtype="double")
    fp = 'Liu_processed/1.png'
    img_gray = cv2.imread(fp, cv2.IMREAD_GRAYSCALE)
    ret, img_binary = cv2.threshold(img_gray, 127, 1, cv2.THRESH_BINARY_INV)
    sigma = 0.9*n**(-0.52) # select scale parameter sigma
    F = np.zeros((k,k), dtype="double")
    Q = discrete_hermite_Q(n,k,sigma) # Q is the storage for Guassian_Hermite polynominal, shape = (n+1,k) <-- (p,i).  p [0,n] and i [0,k-1]
    F = img_binary.T
    M_max = np.dot(np.dot(Q,F),Q.T)
    M_min = np.dot(np.dot(Q,F),Q.T)

    for i in range(0,fileCount):
        img_gray = cv2.imread(filenames[i], cv2.IMREAD_GRAYSCALE)
        ret, img_binary = cv2.threshold(img_gray, 127, 1, cv2.THRESH_BINARY_INV)
        Q = discrete_hermite_Q(n,k,sigma)
        F = img_binary.T
        M = np.dot(np.dot(Q,F),Q.T)
        M_max = np.maximum(M,M_max)
        M_min = np.minimum(M,M_min)
    return M_max, M_min

```

```python
k =256; n =31
M,_  =  M_range(k,n,filenames,fileCount=1)
M_max,M_min = M_range(k,n,filenames,fileCount)
```

## Distribution of M[i,j] for All Characters

```python
from scipy.stats import norm
def M_distribution(k, n, filenames, fileCount,p,q):
    M_pq = np.zeros(fileCount, dtype="double")
    sigma = 0.9*n**(-0.52) # select scale parameter sigma
    F = np.zeros((k,k), dtype="double")
    Q = discrete_hermite_Q(n,k,sigma) # Q is the storage for Guassian_Hermite polynominal, shape = (n+1,k) <-- (p,i).  p [0,n] and i [0,k-1]
    Q_p = Q[p]
    Q_q = Q[q]

    for file in range(0,fileCount):
        img_gray = cv2.imread(filenames[file], cv2.IMREAD_GRAYSCALE)
        ret, img_binary = cv2.threshold(img_gray, 127, 1, cv2.THRESH_BINARY_INV)
        F = img_binary.T
        M_temp = np.dot(np.dot(Q_p,F),Q_q.T)
        M_pq[file] = M_temp
    return M_pq

def draw_hist(M,p,q):
    data = M
    mu = np.mean(data)
    sigma = np.std(data)
    bins = np.linspace(min(data),max(data),20)
    plt.hist(data,bins,density=1, alpha=0.75)
    y = norm.pdf(bins, mu, sigma)
    
    plt.grid(True)
    plt.plot(bins, y, 'r--') #y
    plt.xlabel('Value of M[{0},{1}]'.format(p,q))
    plt.ylabel('Frequency')
    plt.title('Frequency Distribution of M[{0},{1}] for All Characters'.format(p,q))
    plt.show()
```

```python
k =256; n =31
M = np.zeros((n+1,n+1,fileCount), dtype="double")
p = 10; q = 10
if all(M[p,q]) == 0:
    M[p,q] = M_distribution(k,n,filenames, fileCount,p,q)
print(M[p,q])
draw_hist(M[p,q],p,q)
```

```python
for p in range(0,30,5):
    for q in range(0,30,5):
        if all(M[p,q]) == 0:
            M[p,q] = M_distribution(k,n,filenames, fileCount,p,q)
        draw_hist(M[p,q],p,q)
```

```python
import ipywidgets as widgets
from IPython.display import display, HTML
javascript_functions = {False: "hide()", True: "show()"}
button_descriptions  = {False: "Show code", True: "Hide code"}
def toggle_code(state):
    output_string = "<script>$(\"div.input\").{}</script>"
    output_args   = (javascript_functions[state],)
    output        = output_string.format(*output_args)
    display(HTML(output))
def button_action(value):
    state = value.new
    toggle_code(state)
    value.owner.description = button_descriptions[state]
state = False
toggle_code(state)
button = widgets.ToggleButton(state, description = button_descriptions[state])
button.observe(button_action, "value")
display(button)
```
